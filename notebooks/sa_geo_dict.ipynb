# Simple notebook to convert excel sheet of SA Geology codes to JSON

```python
import numpy as np
import pandas as pd
import json
```

```python
path = 'C:\\Users\\dutchr01\\MyPyScripts\\Mason_pet_extract\\sa_geo_codes.xlsx'
df = pd.read_excel(path,sheet_name=2, header = None,index_col=0)
```

```python
df
```

```python
df[1] = df[1].str.replace('META-','META').str.replace('-',' ').str.lower()
```

```python
df
```

```python
rock_code_dict = df.to_dict()
```

```python
new = rock_code_dict[1]
```

```python
print(new)
```

```python
json.dump(new,open('rock_dict.json','w'))
```
